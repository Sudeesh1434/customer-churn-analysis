```python
import pandas as pd 
```

```python
df=pd.read_csv('customer_churn_project_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['InternetService']=df['InternetService'].fillna('unknown')
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=df, x="Churn")
plt.title('cutomer distribution')
```

```python
sns.countplot(data=df,x='gender',hue='Churn')
plt.title('gender vs Churn')
```

```python
sns.histplot(df['MonthlyCharges'],kde=True)
plt.title('Monthly Charges Distribution')
plt.xlabel('MonthlyCharges')
plt.ylabel('count')
```

```python
sns.histplot(data=df, x="MonthlyCharges", hue="Churn", kde=True)
plt.title("Monthly Charges vs Churn")
```

```python
sns.countplot(data=df,x='InternetService',hue='Churn')
plt.title('Internet Service vs Churn')
plt.xticks(rotation=45)
```

```python
sns.boxplot(data=df,x='Churn',hue='tenure')
plt.title('Churn vs tenure')
plt.xlabel('Churn')
plt.ylabel('tenure(Month)')
```

```python
sns.boxplot(data=df,x='Churn',hue='MonthlyCharges')
plt.title('Monthly Charges vs Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
```

```python
sns.countplot(data=df,x='Contract',hue='Churn')
plt.title('Contract Type vs Churn')
plt.xlabel('Contract Type')
plt.ylabel('Customer count')
```

```python
sns.boxplot(data=df,x='Churn',y='tenure')
plt.title('Tenure vs Churn')
plt.xlabel('Churn')
plt.ylabel('Tenure (Months)')
```
